```python
X_log = np.log1p(Xp) 
```

```python
sns.boxplot(X_log)
plt.xticks(rotation=90)    
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler() 

X_scaled = scaler.fit_transform(X_log)
```

```python
from sklearn.decomposition import PCA   

pca = PCA(n_components=2) 

X_s = pca.fit_transform(X_scaled)
```

```python
X_f_df = pd.DataFrame(X_s,columns=["PCA1","PCA2"])
```

```python
sns.scatterplot(X_f_df,x="PCA1",y="PCA2",palette="Paired")
```

```python
sns.heatmap(X_f_df.corr(),annot=True,fmt=".1f")
```

```python
sns.boxplot(X_f_df)
```

```python
from sklearn.cluster import KMeans
```

```python
liste_inertias = []

for i in range(1,12):
    model = KMeans(n_clusters = i,init="k-means++",random_state=42)

    model_ = model.fit(X_f_df)

    liste_inertias.append(model_.inertia_)

liste_inertias
```

```python
sns.lineplot(x=range(1,12),y=liste_inertias,marker="o",color="green")
plt.xlabel('Number of clusters')
plt.ylabel("Inertias")
```

```python
from kneed import KneeLocator

best_k = KneeLocator(range(1,12),liste_inertias,curve="convex",direction="decreasing").elbow

best_k
```

```python
model_kmeans = KMeans(n_clusters=5,init="k-means++",random_state=1)

y_model = model_kmeans.fit(X_f_df)

y_model
```

```python
centroides = pd.DataFrame(model_kmeans.cluster_centers_)

centroides
```

```python
y_pred = y_model.predict(X_f_df)

from sklearn.metrics import silhouette_score

score = silhouette_score(X_f_df,y_pred)

print(score)
```

```python
X_f_df['clusters'] = y_pred

X_f_df.head()
```

```python
plt.figure(figsize=(11,5))
sns.scatterplot(data=X_f_df,x='PCA1',y='PCA2',hue='clusters',palette="Paired",s=60)
sns.scatterplot(data=centroides,x=0,y=1,color='r',label ='centroides',s=500)
```

```python

loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

plt.figure(figsize=(14, 10))

sns.scatterplot(data=X_f_df, x='PCA1', y='PCA2', hue='clusters', palette="Paired", alpha=0.5, s=50)


scale_factor = 3  

for i, feature in enumerate(X.columns):
    plt.arrow(0, 0, loadings[i, 0] * scale_factor, loadings[i, 1] * scale_factor, 
              color='red', alpha=0.8, head_width=0.1)
    

    plt.text(loadings[i, 0] * scale_factor * 1.1, loadings[i, 1] * scale_factor * 1.1, 
             feature, color='darkgreen', fontweight='bold', fontsize=10)

plt.xlabel('PC1 (Spending Volume & Credit Limit)')
plt.ylabel('PC2 (Cash Advance & Debt)')
plt.title('Credit Card Biplot: Customer Segments vs Financial Features')
plt.axhline(0, color='black', linestyle='--', alpha=0.3)
plt.axvline(0, color='black', linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.2)
plt.show()
```

## Results' Interpretation

Credit Card Customer Segmentation: Interpretation

The Biplot reveals that credit card usage is primarily driven by two axes: Spending Intensity (PC1) and Debt/Cash reliance (PC2).
The 5 Strategic Clusters

    Cluster 1 (Blue - Top Right) | The "Prime Transactors"

        Behavior: High scores in PURCHASES and PRC_FULL_PAYMENT.

        Insight: These are your most valuable customers. They spend heavily but also pay their balance in full. They use the card for the rewards and convenience, not as a loan.

    Cluster 2 (Green - Middle Right) | The "Active Installment Shoppers"

        Behavior: Heavily influenced by INSTALLMENTS_PURCHASES and PURCHASES_FREQUENCY.

        Insight: These customers use the card to manage their monthly budget. They buy often but prefer spreading the costs over time.

    Cluster 4 (Red - Far Left) | The "Passive Users"

        Behavior: Located far away from all major spending arrows.

        Insight: These are low-activity or inactive accounts. They maintain low balances and rarely use the card.

    Cluster 3 (Light Green - Bottom Left) | The "Cash-Reliant Borrowers"

        Behavior: Pulled strongly by CASH_ADVANCE, CASH_ADVANCE_TRX, and BALANCE.

        Insight: This is a high-risk/high-revenue group. They use the card primarily to withdraw cash and carry a significant debt balance month-to-month.

    Cluster 0 (Light Blue - Middle) | The "Standard Revolvers"

        Behavior: High BALANCE_FREQUENCY.

        Insight: These are average users who keep their card active and occasionally carry a balance but don't have the extreme spending of the VIPs or the high cash-withdrawal rates of the borrowers.
